##### Comment
It may be interesting to see the intake and outcome of dogs depending on their size. To do this I am creating a categorical variable 'size' where I am sorting the dogs into 'teacup', 'toy', 'small', 'medium', 'large', 'giant' according to their average weight. The categories for smaller sizes will also include puppies of all breeds.

I am using the information on weight found on this page:https://pettr.app/blog/dog-size-guide/

```python
avg_weight=((dogs['breed_max_weight_male']+dogs['breed_min_weight_male'])/2+(
    dogs['breed_max_weight_female']+dogs['breed_min_weight_female'])/2)/2

for i,w in enumerate(avg_weight):
    if w<= 5:
        dogs.loc[i,'size']= 'teacup'
    elif w <= 12:
        dogs.loc[i,'size']= 'toy'
    elif w<= 24:
        dogs.loc[i,'size']= 'small'
    elif w <= 59:
        dogs.loc[i,'size']= 'medium'
    elif w<= 99:
        dogs.loc[i,'size']= 'large'
    else:
        dogs.loc[i,'size']= 'giant'
```

```python
intake_size=pd.crosstab(dogs['size'],dogs['intake_type'])
norm_is=intake_size.div(intake_size.sum(1).astype(float),axis=0)
norm_is.plot(kind='bar', stacked=True, figsize=(10,6), colormap='tab20')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Intake type per size')
plt.grid()
plt.show()
```

```python
intake_size=pd.crosstab(dogs['size'],dogs['outcome_type'])
norm_is=intake_size.div(intake_size.sum(1).astype(float),axis=0)
norm_is.plot(kind='bar', stacked=True, figsize=(10,6), colormap='tab20')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Outcome type per size')
plt.grid()
plt.show()
```

##### Comment
There are still several missing values in the last 12 categories.
To replace them I am first of all grouping by breed, then I am replacing each with a value randomly extracted from a normal distribution with mean and standard deviation relative to the breed for each category.

```python
def replace0(x):
    mu=x.mean()
    std=x.std()
    attr=np.random.normal(mu, std, len(x))
    for g,h in enumerate(attr):
        if h<=0:
            attr[g]=mu
    return attr
```

##### Comment
From the crosstab plot we can see that the only two outcomes subtypes for 'Adoption' are 'Foster' and 
'Offsite'. To deal with all the NaN in the 'Adoption' category I am going to substitute them by random sampling with a weighted outcome based on the percentage of 'Adoption' and 'foster' subtypes in the data.


```python
foster=len(dogs_step3[dogs_step3['outcome_subtype']=='Foster']['outcome_subtype'])
offsite=len(dogs_step3[dogs_step3['outcome_subtype']=='Offsite']['outcome_subtype'])
pct_f=foster/(foster+offsite)*100
pct_o=offsite/(foster+offsite)*100
n_missing=len(dogs_step3[dogs_step3['outcome_type']=='Adoption'])-(foster+offsite)
```

```python
from random import choices
outcomes=['Foster', 'Offsite']
foof=choices(outcomes, weights=[pct_f,pct_o], k=len(dogs))
dogs_step3['outcome_subtype']=dogs_step3['outcome_subtype'].fillna(pd.Series(foof))
```

##### Comment

From the correlation matrix (at the very start of the notebook) we can see that the columns 'breed_min_life_expectancy' and 'breed_max_life_expectancy' are strongly correlated; to summarise the information they contain I am combining them in the columns 'breed_avg_life_expectancy'.

Since the 'lenght_of_stay' and 'age' columns are in days, I am converting the average life expectancy to days as well (currently in years).

```python
dogs_step4['breed_avg_life_expectancy']=(dogs_step4['breed_max_life_expectancy']+
                                         dogs_step4['breed_min_life_expectancy'])/2

for l,y in enumerate(dogs_step4['breed_avg_life_expectancy']):
    years, fraction = divmod(y,1)
    days = round(fraction*365.2425)
    dogs_step4.loc[l,'breed_avg_life_expectancy']= int(years*365 + days)
```

```python
dogs_step4=dogs_step4.drop(['breed_min_life_expectancy','breed_max_life_expectancy'], axis=1)
```

##### Comment

The height and weights columns are very strongly correlated. Moreover these averages across males and females are correlated.

To reduce the dimensionality of the dataset I have chosen 'height' as a variable able to capture also the variable 'weight'. 

I am replacing all the breed averages columns with a single column called 'breed_avg_weight'. To obtain this column I am calculating the average between min and max height and also the average between male and female.

```python
dogs_step4['breed_avg_height']=((
    (dogs_step4['breed_max_height_male']+dogs_step4['breed_min_height_male'])/2)+(
        dogs_step4['breed_max_height_female']+dogs_step4['breed_min_height_female'])/2)/2
```

##### Comment 
There are more than 100 different values in the 'color' column for dogs having their fur of a single colour, two different colours and even one or two colours with a pattern.

One hot encoding of this attribute would add a number of columns to the dataset that, in my opinion, would not add enough significant information to justify adding them.

To simplify the attribute I have made the assumption that the first colour (if more than one is present) is the dominant one and I have ignored the patterns.

```python
for o,p in enumerate(dogs_step4['color']):
    if ' ' in p:
        p=p.split(' ')[0]
    if '/' in p:
        dogs_step4.loc[o,'color']= p.split('/')[0]
    else:
        dogs_step4.loc[o,'color']= p
np.unique(dogs_step4['color'])
```
